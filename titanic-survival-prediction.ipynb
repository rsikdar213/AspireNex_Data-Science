```python
data = data.select_dtypes(include=['float64','int64'])
```

```python
data = data.drop(columns=['PassengerId'])
```

```python
data['Fare'].fillna(data['Fare'].mean(), inplace=True)
```

```python
X = data.drop(columns=['Survived'])
y = data['Survived']
```

```python
X.describe()
```

```python
mlp_clf = MLPClassifier(solver='adam',learning_rate="adaptive",alpha=1e-3,validation_fraction=0.1,random_state=1,max_iter=500,tol=1e-5, hidden_layer_sizes=(64,32,16,8,4,2))
```

```python
lr_clf = LogisticRegression(solver='liblinear', max_iter=400)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
y_train.shape
```

```python
mlp_clf.fit(X_train,y_train)
```

```python
lr_clf.fit(X_train, y_train)
```

```python
# y_pred = mlp_clf.predict(X_test)
y_pred = lr_clf.predict(X_test)
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
